```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
class Particle():
    def __init__(self,x,y,vx,vy,color):
        self.x = x
        self.y = y
        self.vx = vx
        self.vy = vy
        self.color = color
    def positionChange(self,dt):
        self.x = self.x + self.vx*dt
        self.y = self.y + self.vy*dt
        if (self.x==0 or self.x==10):
            self.vx = -self.vx
        elif (self.x<0):
            self.x = -self.x
            self.vx = -self.vx
        elif (self.x>10):
            self.x = 10 - self.x
            self.vx = -self.vx

        if (self.y==0 or self.y==10):
            self.vy = -self.vy
        elif (self.y<0):
            self.y = -self.y
            self.vy = -self.vy
        elif (self.y>10):
            self.y = 10 - self.y
            self.vy = -self.vy

```

```python
N=50
par_array = []
```

```python
for i in range(0,N):
    x = random.random()*10
    y = random.random()*10
    vx = random.random()*10-10
    vy = random.random()*10-10
    RGB = (random.random(),random.random(),random.random())
    par_array.append(Particle(x,y,vx,vy,RGB))
```

```python
def timestamp(par_array):    
    plt.figure()
    for i in range(0,50):
        plt.scatter(par_array[i].x,par_array[i].y)
```

```python
dt = 0.2; times=10
```

```python
for j in range(0,times):
    for i in range(0,N):
        par_array[i].positionChange(dt)
    timestamp(par_array)
```
